```python
# %% [markdown]
# # Há»‡ thá»‘ng PhÃ¢n tÃ­ch & Tá»‘i Æ°u Chiáº¿n lÆ°á»£c Há»c táº­p vá»›i ML vÃ  Reinforcement Learning
# 
# ## Má»¥c tiÃªu:
# 1. **PhÃ¢n cá»¥m** sinh viÃªn dá»±a trÃªn hÃ nh vi há»c táº­p
# 2. **TrÃ­ch xuáº¥t** ma tráº­n xÃ¡c suáº¥t Markov tá»« má»—i cá»¥m
# 3. **MÃ´ phá»ng** dá»¯ liá»‡u hÃ nh vi há»c táº­p
# 4. **Huáº¥n luyá»‡n** Q-Learning Ä‘á»ƒ tá»‘i Æ°u chiáº¿n lÆ°á»£c

# %% [markdown]
# ## 1. Import Libraries & Load Data

# %%
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score
import json
import warnings
warnings.filterwarnings('ignore')

# Cáº¥u hÃ¬nh hiá»ƒn thá»‹
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("âœ“ Libraries imported successfully")

import json
with open("../data/features_scaled_report.json", 'r', encoding='utf-8') as f:
    data = json.load(f)
df = pd.DataFrame(data)

print(f"âœ“ Loaded {len(df)} students")
print(f"âœ“ Features: {len(df.columns)} columns")
df.head()

# %% [markdown]
# ## 2. Data Preprocessing & Feature Engineering

# %%
# Lá»±a chá»n features quan trá»ng cho phÃ¢n cá»¥m
feature_cols = [col for col in df.columns if col != 'userid']

# Táº¡o thÃªm má»™t sá»‘ features tá»•ng há»£p
df['engagement_score'] = (df['viewed'] + df['submitted'] + df['created']) / 3
df['assignment_completion'] = df['\\mod_assign\\event\\assessable_submitted']
df['quiz_participation'] = (df['\\mod_quiz\\event\\attempt_started'] + 
                             df['\\mod_quiz\\event\\attempt_submitted']) / 2

# Cáº­p nháº­t feature_cols
feature_cols = [
    'total_events',
    '\\assignsubmission_file\\event\\assessable_uploaded',
    '\\mod_assign\\event\\assessable_submitted',
    '\\mod_assign\\event\\course_module_viewed',
    '\\mod_quiz\\event\\attempt_started',
    '\\mod_quiz\\event\\attempt_submitted',
    'viewed',
    'submitted',
    'created',
    'mean_module_grade',
    'engagement_score',
    'assignment_completion',
    'quiz_participation'
]

X = df[feature_cols].values

# Chuáº©n hÃ³a dá»¯ liá»‡u
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"âœ“ Features selected: {len(feature_cols)}")
print(f"âœ“ Data shape: {X_scaled.shape}")
print(f"\nðŸ“Š Feature Statistics:")
print(pd.DataFrame(X, columns=feature_cols).describe())

# %% [markdown]
# ## 3. Determining Optimal Number of Clusters

# %%
# Elbow Method & Silhouette Score
inertias = []
silhouette_scores = []
db_scores = []
K_range = range(2, 8)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
    db_scores.append(davies_bouldin_score(X_scaled, kmeans.labels_))

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Elbow plot
axes[0].plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[0].set_ylabel('Inertia', fontsize=12)
axes[0].set_title('Elbow Method', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Silhouette score
axes[1].plot(K_range, silhouette_scores, 'go-', linewidth=2, markersize=8)
axes[1].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[1].set_ylabel('Silhouette Score', fontsize=12)
axes[1].set_title('Silhouette Analysis', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Davies-Bouldin score (lower is better)
axes[2].plot(K_range, db_scores, 'ro-', linewidth=2, markersize=8)
axes[2].set_xlabel('Number of Clusters (k)', fontsize=12)
axes[2].set_ylabel('Davies-Bouldin Score', fontsize=12)
axes[2].set_title('Davies-Bouldin Index', fontsize=14, fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

optimal_k = 3  # CÃ³ thá»ƒ Ä‘iá»u chá»‰nh dá»±a trÃªn biá»ƒu Ä‘á»“
print(f"\nâœ“ Recommended optimal k = {optimal_k}")

# %% [markdown]
# ## 4. KMeans Clustering

# %%
# Thá»±c hiá»‡n KMeans vá»›i k tá»‘i Æ°u
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=20)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Thá»‘ng kÃª cÃ¡c cá»¥m
print("=" * 60)
print("CLUSTER STATISTICS")
print("=" * 60)
for i in range(optimal_k):
    cluster_data = df[df['cluster'] == i]
    print(f"\nðŸŽ¯ CLUSTER {i}: {len(cluster_data)} students ({len(cluster_data)/len(df)*100:.1f}%)")
    print(f"   Average Grade: {cluster_data['mean_module_grade'].mean():.3f}")
    print(f"   Average Events: {cluster_data['total_events'].mean():.3f}")
    print(f"   Engagement Score: {cluster_data['engagement_score'].mean():.3f}")
    print(f"   Quiz Participation: {cluster_data['quiz_participation'].mean():.3f}")

# %% [markdown]
# ## 5. Cluster Visualization

# %%
# PCA cho visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Táº¡o DataFrame cho visualization
viz_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'Cluster': df['cluster'],
    'Grade': df['mean_module_grade'],
    'UserID': df['userid']
})

# Scatter plot
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Clusters
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']
for i in range(optimal_k):
    cluster_data = viz_df[viz_df['Cluster'] == i]
    axes[0].scatter(cluster_data['PC1'], cluster_data['PC2'], 
                   c=colors[i], label=f'Cluster {i}', 
                   s=100, alpha=0.6, edgecolors='black', linewidth=1)

axes[0].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
axes[0].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
axes[0].set_title('Student Clusters (PCA)', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot 2: Grade heatmap
scatter = axes[1].scatter(viz_df['PC1'], viz_df['PC2'], 
                         c=viz_df['Grade'], cmap='RdYlGn',
                         s=100, alpha=0.6, edgecolors='black', linewidth=1)
axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)', fontsize=12)
axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)', fontsize=12)
axes[1].set_title('Student Performance Heatmap', fontsize=14, fontweight='bold')
plt.colorbar(scatter, ax=axes[1], label='Grade')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 6. Cluster Profiling vá»›i Radar Chart

# %%
# Táº¡o profile cho má»—i cluster
cluster_profiles = {}
profile_features = [
    'total_events', 'viewed', 'submitted', 'created',
    'assignment_completion', 'quiz_participation', 'mean_module_grade'
]

for i in range(optimal_k):
    cluster_data = df[df['cluster'] == i]
    profile = cluster_data[profile_features].mean()
    cluster_profiles[f'Cluster {i}'] = profile

profile_df = pd.DataFrame(cluster_profiles).T

# Radar chart
fig, axes = plt.subplots(1, optimal_k, figsize=(6*optimal_k, 5), subplot_kw=dict(projection='polar'))
if optimal_k == 1:
    axes = [axes]

angles = np.linspace(0, 2 * np.pi, len(profile_features), endpoint=False).tolist()
angles += angles[:1]

for idx, (cluster_name, row) in enumerate(profile_df.iterrows()):
    values = row.tolist()
    values += values[:1]
    
    axes[idx].plot(angles, values, 'o-', linewidth=2, color=colors[idx])
    axes[idx].fill(angles, values, alpha=0.25, color=colors[idx])
    axes[idx].set_xticks(angles[:-1])
    axes[idx].set_xticklabels(profile_features, size=8)
    axes[idx].set_ylim(0, 1)
    axes[idx].set_title(f'{cluster_name}\n({len(df[df["cluster"]==idx])} students)', 
                       fontsize=12, fontweight='bold', pad=20)
    axes[idx].grid(True)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 7. Markov Chain - Ma tráº­n XÃ¡c suáº¥t Chuyá»ƒn tráº¡ng thÃ¡i

# %%
# Äá»‹nh nghÄ©a cÃ¡c states (events) quan trá»ng
states = [
    'view_course',
    'view_assignment', 
    'submit_assignment',
    'attempt_quiz',
    'submit_quiz',
    'view_resource'
]

def create_markov_matrix(cluster_df):
    """
    Táº¡o ma tráº­n chuyá»ƒn Ä‘á»•i Markov tá»« dá»¯ liá»‡u cluster
    """
    n_states = len(states)
    transition_matrix = np.zeros((n_states, n_states))
    
    # TÃ­nh xÃ¡c suáº¥t chuyá»ƒn Ä‘á»•i dá»±a trÃªn táº§n suáº¥t events
    for idx in cluster_df.index:
        student = cluster_df.loc[idx]
        
        # XÃ¡c suáº¥t chuyá»ƒn tá»« view_course
        if student['\\mod_assign\\event\\course_module_viewed'] > 0:
            transition_matrix[0][1] += student['\\mod_assign\\event\\course_module_viewed']
            transition_matrix[0][5] += student['viewed']
        
        # XÃ¡c suáº¥t chuyá»ƒn tá»« view_assignment Ä‘áº¿n submit
        if student['\\mod_assign\\event\\course_module_viewed'] > 0:
            transition_matrix[1][2] += student['\\mod_assign\\event\\assessable_submitted']
            transition_matrix[1][0] += 0.2  # Quay láº¡i xem course
        
        # XÃ¡c suáº¥t sau submit assignment
        if student['\\mod_assign\\event\\assessable_submitted'] > 0:
            transition_matrix[2][3] += student['\\mod_quiz\\event\\attempt_started'] * 0.5
            transition_matrix[2][5] += 0.3
            transition_matrix[2][0] += 0.2
        
        # XÃ¡c suáº¥t chuyá»ƒn tá»« attempt quiz
        if student['\\mod_quiz\\event\\attempt_started'] > 0:
            transition_matrix[3][4] += student['\\mod_quiz\\event\\attempt_submitted']
            transition_matrix[3][3] += 0.1  # Thá»­ láº¡i
        
        # XÃ¡c suáº¥t sau submit quiz
        if student['\\mod_quiz\\event\\attempt_submitted'] > 0:
            transition_matrix[4][0] += 0.4
            transition_matrix[4][5] += 0.3
            transition_matrix[4][1] += 0.3
        
        # XÃ¡c suáº¥t tá»« view resource
        transition_matrix[5][0] += 0.3
        transition_matrix[5][1] += 0.4
        transition_matrix[5][5] += 0.3
    
    # Chuáº©n hÃ³a thÃ nh xÃ¡c suáº¥t
    row_sums = transition_matrix.sum(axis=1, keepdims=True)
    row_sums[row_sums == 0] = 1  # TrÃ¡nh chia cho 0
    transition_matrix = transition_matrix / row_sums
    
    return transition_matrix

# Táº¡o Markov matrix cho má»—i cluster
markov_matrices = {}
for i in range(optimal_k):
    cluster_data = df[df['cluster'] == i]
    markov_matrices[i] = create_markov_matrix(cluster_data)
    
    print(f"\n{'='*60}")
    print(f"MARKOV TRANSITION MATRIX - CLUSTER {i}")
    print(f"{'='*60}")
    markov_df = pd.DataFrame(markov_matrices[i], 
                            index=states, 
                            columns=states)
    print(markov_df.round(3))

# %% [markdown]
# ## 8. Visualize Markov Chains

# %%
# Heatmap cho má»—i Markov matrix
fig, axes = plt.subplots(1, optimal_k, figsize=(8*optimal_k, 6))
if optimal_k == 1:
    axes = [axes]

for i in range(optimal_k):
    sns.heatmap(markov_matrices[i], 
                annot=True, 
                fmt='.2f', 
                cmap='YlOrRd',
                xticklabels=states,
                yticklabels=states,
                cbar_kws={'label': 'Transition Probability'},
                ax=axes[i],
                vmin=0,
                vmax=1)
    axes[i].set_title(f'Cluster {i} - Markov Chain\n({len(df[df["cluster"]==i])} students)', 
                     fontsize=12, fontweight='bold')
    axes[i].set_xlabel('To State', fontsize=10)
    axes[i].set_ylabel('From State', fontsize=10)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 9. Simulation - MÃ´ phá»ng HÃ nh vi Há»c táº­p

# %%
def simulate_learning_path(transition_matrix, initial_state, n_steps=20):
    """
    MÃ´ phá»ng quÃ¡ trÃ¬nh há»c táº­p cá»§a sinh viÃªn dá»±a trÃªn Markov chain
    """
    current_state = initial_state
    path = [current_state]
    
    for _ in range(n_steps):
        # Chá»n state tiáº¿p theo dá»±a trÃªn xÃ¡c suáº¥t
        probabilities = transition_matrix[current_state]
        next_state = np.random.choice(len(states), p=probabilities)
        path.append(next_state)
        current_state = next_state
    
    return path

# MÃ´ phá»ng cho má»—i cluster
n_simulations = 100
n_steps = 30

simulation_results = {}

for cluster_id in range(optimal_k):
    print(f"\n{'='*60}")
    print(f"SIMULATING CLUSTER {cluster_id}")
    print(f"{'='*60}")
    
    simulated_paths = []
    for _ in range(n_simulations):
        # Báº¯t Ä‘áº§u tá»« state 0 (view_course)
        path = simulate_learning_path(markov_matrices[cluster_id], 0, n_steps)
        simulated_paths.append(path)
    
    simulation_results[cluster_id] = simulated_paths
    
    # PhÃ¢n tÃ­ch paths
    all_states = [state for path in simulated_paths for state in path]
    state_counts = pd.Series(all_states).value_counts().sort_index()
    
    print(f"\nState Distribution trong {n_simulations} simulations:")
    for state_idx, count in state_counts.items():
        print(f"  {states[state_idx]:<20}: {count:>5} láº§n ({count/len(all_states)*100:>5.1f}%)")

# Visualize simulation results
fig, axes = plt.subplots(1, optimal_k, figsize=(8*optimal_k, 6))
if optimal_k == 1:
    axes = [axes]

for cluster_id in range(optimal_k):
    # Äáº¿m táº§n suáº¥t cÃ¡c states
    all_states = [state for path in simulation_results[cluster_id] for state in path]
    state_counts = pd.Series(all_states).value_counts().sort_index()
    
    axes[cluster_id].bar(range(len(states)), 
                        [state_counts.get(i, 0) for i in range(len(states))],
                        color=colors[cluster_id],
                        alpha=0.7,
                        edgecolor='black')
    axes[cluster_id].set_xticks(range(len(states)))
    axes[cluster_id].set_xticklabels(states, rotation=45, ha='right')
    axes[cluster_id].set_ylabel('Frequency', fontsize=11)
    axes[cluster_id].set_title(f'Cluster {cluster_id} - State Distribution\n({n_simulations} simulations)', 
                              fontsize=12, fontweight='bold')
    axes[cluster_id].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 10. Visualize Sample Learning Paths

# %%
# Váº½ má»™t sá»‘ learning paths máº«u
fig, axes = plt.subplots(optimal_k, 1, figsize=(14, 5*optimal_k))
if optimal_k == 1:
    axes = [axes]

for cluster_id in range(optimal_k):
    # Láº¥y 5 paths ngáº«u nhiÃªn Ä‘á»ƒ váº½
    sample_paths = np.random.choice(len(simulation_results[cluster_id]), 
                                   min(5, len(simulation_results[cluster_id])), 
                                   replace=False)
    
    for idx, path_idx in enumerate(sample_paths):
        path = simulation_results[cluster_id][path_idx]
        axes[cluster_id].plot(range(len(path)), path, 
                             marker='o', alpha=0.6, linewidth=2,
                             label=f'Student {idx+1}')
    
    axes[cluster_id].set_xlabel('Step', fontsize=11)
    axes[cluster_id].set_ylabel('State', fontsize=11)
    axes[cluster_id].set_yticks(range(len(states)))
    axes[cluster_id].set_yticklabels(states)
    axes[cluster_id].set_title(f'Cluster {cluster_id} - Sample Learning Paths', 
                              fontsize=12, fontweight='bold')
    axes[cluster_id].legend(loc='upper right')
    axes[cluster_id].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 11. Q-Learning - Reinforcement Learning cho Chiáº¿n lÆ°á»£c Tá»‘i Æ°u

# %%
class QLearningAgent:
    """
    Q-Learning Agent Ä‘á»ƒ tÃ¬m chiáº¿n lÆ°á»£c há»c táº­p tá»‘i Æ°u
    """
    def __init__(self, n_states, n_actions, learning_rate=0.1, discount=0.95, epsilon=0.1):
        self.q_table = np.zeros((n_states, n_actions))
        self.lr = learning_rate
        self.gamma = discount
        self.epsilon = epsilon
        self.n_states = n_states
        self.n_actions = n_actions
    
    def choose_action(self, state):
        """Epsilon-greedy action selection"""
        if np.random.random() < self.epsilon:
            return np.random.randint(self.n_actions)
        else:
            return np.argmax(self.q_table[state])
    
    def learn(self, state, action, reward, next_state):
        """Update Q-table"""
        current_q = self.q_table[state, action]
        max_next_q = np.max(self.q_table[next_state])
        new_q = current_q + self.lr * (reward + self.gamma * max_next_q - current_q)
        self.q_table[state, action] = new_q
    
    def get_policy(self):
        """Extract optimal policy from Q-table"""
        return np.argmax(self.q_table, axis=1)

def reward_function(state, next_state, cluster_grade_avg):
    """
    Äá»‹nh nghÄ©a hÃ m reward dá»±a trÃªn:
    - Tiáº¿n Ä‘á»™ há»c táº­p (submit assignment, quiz)
    - Alignment vá»›i cluster cÃ³ performance cao
    """
    rewards = {
        'view_course': 0.1,
        'view_assignment': 0.2,
        'submit_assignment': 0.8,
        'attempt_quiz': 0.5,
        'submit_quiz': 1.0,
        'view_resource': 0.3
    }
    
    base_reward = rewards.get(states[next_state], 0)
    
    # Bonus náº¿u cluster cÃ³ grade cao
    grade_bonus = cluster_grade_avg * 0.5
    
    # Penalty náº¿u láº·p láº¡i state quÃ¡ nhiá»u (khÃ´ng tiáº¿n bá»™)
    if state == next_state:
        base_reward -= 0.2
    
    return base_reward + grade_bonus

# Train Q-Learning agent cho má»—i cluster
q_agents = {}
n_episodes = 1000
episode_length = 30

print("\n" + "="*60)
print("TRAINING Q-LEARNING AGENTS")
print("="*60)

for cluster_id in range(optimal_k):
    print(f"\nðŸŽ¯ Training Agent for Cluster {cluster_id}...")
    
    # Láº¥y average grade cá»§a cluster
    cluster_grade = df[df['cluster'] == cluster_id]['mean_module_grade'].mean()
    
    # Khá»Ÿi táº¡o agent
    agent = QLearningAgent(n_states=len(states), 
                          n_actions=len(states),
                          learning_rate=0.1,
                          discount=0.95,
                          epsilon=0.2)
    
    # Training loop
    total_rewards = []
    
    for episode in range(n_episodes):
        state = 0  # Báº¯t Ä‘áº§u tá»« view_course
        episode_reward = 0
        
        for step in range(episode_length):
            # Chá»n action
            action = agent.choose_action(state)
            
            # Thá»±c hiá»‡n action (chuyá»ƒn state theo Markov)
            probabilities = markov_matrices[cluster_id][state]
            next_state = action if np.random.random() < probabilities[action] else np.random.choice(len(states), p=probabilities)
            
            # TÃ­nh reward
            reward = reward_function(state, next_state, cluster_grade)
            episode_reward += reward
            
            # Update Q-table
            agent.learn(state, action, reward, next_state)
            
            state = next_state
        
        total_rewards.append(episode_reward)
        
        if (episode + 1) % 200 == 0:
            avg_reward = np.mean(total_rewards[-200:])
            print(f"  Episode {episode+1}/{n_episodes} - Avg Reward: {avg_reward:.2f}")
    
    q_agents[cluster_id] = agent
    
    # Plot learning curve
    plt.figure(figsize=(10, 4))
    plt.plot(total_rewards, alpha=0.3, color=colors[cluster_id])
    plt.plot(pd.Series(total_rewards).rolling(50).mean(), 
            linewidth=2, color=colors[cluster_id],
            label=f'Cluster {cluster_id}')
    plt.xlabel('Episode', fontsize=11)
    plt.ylabel('Total Reward', fontsize=11)
    plt.title(f'Q-Learning Training Progress - Cluster {cluster_id}', 
             fontsize=12, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

print("\nâœ“ Training completed for all clusters!")

# %% [markdown]
# ## 12. Visualize Q-Tables

# %%
# Hiá»ƒn thá»‹ Q-tables
fig, axes = plt.subplots(1, optimal_k, figsize=(8*optimal_k, 6))
if optimal_k == 1:
    axes = [axes]

for cluster_id in range(optimal_k):
    q_table = q_agents[cluster_id].q_table
    
    sns.heatmap(q_table, 
                annot=True, 
                fmt='.2f', 
                cmap='viridis',
                xticklabels=states,
                yticklabels=states,
                cbar_kws={'label': 'Q-Value'},
                ax=axes[cluster_id])
    axes[cluster_id].set_title(f'Cluster {cluster_id} - Q-Table', 
                              fontsize=12, fontweight='bold')
    axes[cluster_id].set_xlabel('Action (Next State)', fontsize=10)
    axes[cluster_id].set_ylabel('Current State', fontsize=10)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 13. Extract Optimal Policies

# %%
print("\n" + "="*60)
print("OPTIMAL LEARNING STRATEGIES (POLICIES)")
print("="*60)

optimal_policies = {}

for cluster_id in range(optimal_k):
    policy = q_agents[cluster_id].get_policy()
    optimal_policies[cluster_id] = policy
    
    print(f"\nðŸŽ¯ CLUSTER {cluster_id} - Optimal Policy:")
    print(f"   (Average Grade: {df[df['cluster'] == cluster_id]['mean_module_grade'].mean():.3f})")
    print(f"   {'='*50}")
    
    for state_idx, action_idx in enumerate(policy):
        print(f"   Khi á»Ÿ: {states[state_idx]:<20} â†’ NÃªn: {states[action_idx]}")

# %% [markdown]
# ## 14. Policy Comparison Visualization

# %%
# Táº¡o heatmap so sÃ¡nh policies
policy_matrix = np.array([optimal_policies[i] for i in range(optimal_k)])

plt.figure(figsize=(12, 4))
sns.heatmap(policy_matrix.T, 
            annot=True, 
            fmt='d',
            cmap='Set3',
            xticklabels=[f'Cluster {i}' for i in range(optimal_k)],
            yticklabels=states,
            cbar_kws={'label': 'Recommended Next State'})
plt.title('Optimal Policy Comparison Across Clusters', 
         fontsize=14, fontweight='bold')
plt.xlabel('Cluster', fontsize=11)
plt.ylabel('Current State', fontsize=11)
plt.tight_layout()
plt.show()

# %% [markdown]
# ## 15. Evaluate Strategies - Simulation vá»›i Optimal Policy

# %%
def simulate_with_policy(policy, transition_matrix, n_steps=30):
    """MÃ´ phá»ng vá»›i optimal policy"""
    state = 0
    path = [state]
    rewards_collected = []
    
    for _ in range(n_steps):
        # Theo policy
        action = policy[state]
        
        # Transition theo Markov
        probabilities = transition_matrix[state]
        if np.random.random() < probabilities[action]:
            next_state = action
        else:
            next_state = np.random.choice(len(states), p=probabilities)
        
        # Thu tháº­p reward
        reward = reward_function(state, next_state, 0.7)  # Giáº£ Ä‘á»‹nh grade trung bÃ¬nh
        rewards_collected.append(reward)
        
        path.append(next_state)
        state = next_state
    
    return path, rewards_collected

# So sÃ¡nh random policy vs optimal policy
print("\n" + "="*60)
print("STRATEGY EVALUATION: Random vs Optimal")
print("="*60)

n_eval_simulations = 100

for cluster_id in range(optimal_k):
    print(f"\nðŸ“Š Cluster {cluster_id}:")
    
    # Random policy
    random_rewards = []
    for _ in range(n_eval_simulations):
        random_policy = np.random.randint(0, len(states), size=len(states))
        _, rewards = simulate_with_policy(random_policy, 
                                         markov_matrices[cluster_id])
        random_rewards.append(sum(rewards))
    
    # Optimal policy
    optimal_rewards = []
    for _ in range(n_eval_simulations):
        _, rewards = simulate_with_policy(optimal_policies[cluster_id],
                                         markov_matrices[cluster_id])
        optimal_rewards.append(sum(rewards))
    
    print(f"   Random Strategy  - Avg Reward: {np.mean(random_rewards):.2f} Â± {np.std(random_rewards):.2f}")
    print(f"   Optimal Strategy - Avg Reward: {np.mean(optimal_rewards):.2f} Â± {np.std(optimal_rewards):.2f}")
    print(f"   Improvement: {(np.mean(optimal_rewards) - np.mean(random_rewards)) / np.mean(random_rewards) * 100:.1f}%")

# %% [markdown]
# ## 16. Comprehensive Visualization

# %%
# Táº¡o dashboard tá»•ng há»£p
fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. Cluster distribution
ax1 = fig.add_subplot(gs[0, 0])
cluster_counts = df['cluster'].value_counts().sort_index()
ax1.bar(cluster_counts.index, cluster_counts.values, 
       color=[colors[i] for i in cluster_counts.index],
       alpha=0.7, edgecolor='black')
ax1.set_xlabel('Cluster')
ax1.set_ylabel('Number of Students')
ax1.set_title('Student Distribution by Cluster', fontweight='bold')
ax1.grid(axis='y', alpha=0.3)

# 2. Average grade by cluster
ax2 = fig.add_subplot(gs[0, 1])
avg_grades = df.groupby('cluster')['mean_module_grade'].mean()
ax2.bar(avg_grades.index, avg_grades.values,
       color=[colors[i] for i in avg_grades.index],
       alpha=0.7, edgecolor='black')
ax2.set_xlabel('Cluster')
ax2.set_ylabel('Average Grade')
ax2.set_title('Performance by Cluster', fontweight='bold')
ax2.set_ylim([0, 1])
ax2.grid(axis='y', alpha=0.3)

# 3. Engagement score by cluster
ax3 = fig.add_subplot(gs[0, 2])
avg_engagement = df.groupby('cluster')['engagement_score'].mean()
ax3.bar(avg_engagement.index, avg_engagement.values,
       color=[colors[i] for i in avg_engagement.index],
       alpha=0.7, edgecolor='black')
ax3.set_xlabel('Cluster')
ax3.set_ylabel('Engagement Score')
ax3.set_title('Engagement by Cluster', fontweight='bold')
ax3.grid(axis='y', alpha=0.3)

# 4-6. Markov matrices
for i in range(min(optimal_k, 3)):
    ax = fig.add_subplot(gs[1, i])
    sns.heatmap(markov_matrices[i], 
               annot=False,
               cmap='YlOrRd',
               cbar=False,
               ax=ax)
    ax.set_title(f'Cluster {i} Markov Chain', fontweight='bold', fontsize=10)
    ax.set_xlabel('')
    ax.set_ylabel('')

# 7-9. Q-tables
for i in range(min(optimal_k, 3)):
    ax = fig.add_subplot(gs[2, i])
    sns.heatmap(q_agents[i].q_table,
               annot=False,
               cmap='viridis',
               cbar=False,
               ax=ax)
    ax.set_title(f'Cluster {i} Q-Table', fontweight='bold', fontsize=10)
    ax.set_xlabel('')
    ax.set_ylabel('')

plt.suptitle('Moodle Learning Analytics Dashboard', 
            fontsize=16, fontweight='bold', y=0.995)
plt.show()

# %% [markdown]
# ## 17. Recommendations for Each Cluster

# %%
print("\n" + "="*70)
print("ðŸ“‹ PERSONALIZED LEARNING RECOMMENDATIONS")
print("="*70)

for cluster_id in range(optimal_k):
    cluster_data = df[df['cluster'] == cluster_id]
    policy = optimal_policies[cluster_id]
    
    print(f"\n{'='*70}")
    print(f"ðŸŽ¯ CLUSTER {cluster_id}: {len(cluster_data)} students")
    print(f"{'='*70}")
    
    # Characteristics
    print(f"\nðŸ“Š Characteristics:")
    print(f"   â€¢ Average Grade: {cluster_data['mean_module_grade'].mean():.2f}")
    print(f"   â€¢ Average Events: {cluster_data['total_events'].mean():.2f}")
    print(f"   â€¢ Engagement: {cluster_data['engagement_score'].mean():.2f}")
    print(f"   â€¢ Quiz Participation: {cluster_data['quiz_participation'].mean():.2f}")
    print(f"   â€¢ Assignment Completion: {cluster_data['assignment_completion'].mean():.2f}")
    
    # Classification
    avg_grade = cluster_data['mean_module_grade'].mean()
    if avg_grade >= 0.7:
        category = "HIGH PERFORMERS ðŸŒŸ"
    elif avg_grade >= 0.4:
        category = "MEDIUM PERFORMERS ðŸ“ˆ"
    else:
        category = "AT-RISK STUDENTS âš ï¸"
    
    print(f"\nðŸ·ï¸  Category: {category}")
    
    # Top 3 recommended actions
    print(f"\nðŸ’¡ Top 3 Recommended Learning Paths:")
    q_table = q_agents[cluster_id].q_table
    
    # TÃ¬m top actions cho má»—i state quan trá»ng
    important_states = [0, 1, 3]  # view_course, view_assignment, attempt_quiz
    for state_idx in important_states:
        top_actions = np.argsort(q_table[state_idx])[-3:][::-1]
        print(f"\n   When '{states[state_idx]}':")
        for rank, action_idx in enumerate(top_actions, 1):
            print(f"      {rank}. â†’ {states[action_idx]} (Q-value: {q_table[state_idx, action_idx]:.2f})")
    
    # Specific advice
    print(f"\nðŸŽ“ Personalized Advice:")
    if avg_grade >= 0.7:
        print("   â€¢ Maintain current study habits")
        print("   â€¢ Focus on challenging materials")
        print("   â€¢ Help peers through forum discussions")
    elif avg_grade >= 0.4:
        print("   â€¢ Increase assignment submission frequency")
        print("   â€¢ Review materials before quizzes")
        print("   â€¢ Engage more with course resources")
    else:
        print("   â€¢ URGENT: Increase course engagement immediately")
        print("   â€¢ Start with viewing all available resources")
        print("   â€¢ Complete assignments before attempting quizzes")
        print("   â€¢ Consider seeking help from instructors")

# %% [markdown]
# ## 18. Export Results

# %%
# Xuáº¥t káº¿t quáº£ ra files
print("\n" + "="*60)
print("ðŸ’¾ EXPORTING RESULTS")
print("="*60)

# 1. Student clusters
df_export = df[['userid', 'cluster', 'mean_module_grade', 'engagement_score', 
                'total_events', 'assignment_completion', 'quiz_participation']]
df_export.to_csv('student_clusters.csv', index=False)
print("âœ“ Saved: student_clusters.csv")

# 2. Markov matrices
for cluster_id in range(optimal_k):
    markov_df = pd.DataFrame(markov_matrices[cluster_id],
                            index=states,
                            columns=states)
    markov_df.to_csv(f'markov_matrix_cluster_{cluster_id}.csv')
    print(f"âœ“ Saved: markov_matrix_cluster_{cluster_id}.csv")

# 3. Q-tables
for cluster_id in range(optimal_k):
    q_df = pd.DataFrame(q_agents[cluster_id].q_table,
                       index=states,
                       columns=states)
    q_df.to_csv(f'q_table_cluster_{cluster_id}.csv')
    print(f"âœ“ Saved: q_table_cluster_{cluster_id}.csv")

# 4. Optimal policies
policy_df = pd.DataFrame({
    'State': states,
    **{f'Cluster_{i}': [states[optimal_policies[i][j]] for j in range(len(states))] 
       for i in range(optimal_k)}
})
policy_df.to_csv('optimal_policies.csv', index=False)
print("âœ“ Saved: optimal_policies.csv")

print("\nâœ“ All results exported successfully!")

# %% [markdown]
# ## 19. Summary Statistics

# %%
print("\n" + "="*70)
print("ðŸ“ˆ FINAL SUMMARY")
print("="*70)

print(f"\nðŸ“Š Dataset Overview:")
print(f"   â€¢ Total Students: {len(df)}")
print(f"   â€¢ Number of Clusters: {optimal_k}")
print(f"   â€¢ Features Used: {len(feature_cols)}")
print(f"   â€¢ States Modeled: {len(states)}")

print(f"\nðŸŽ¯ Clustering Performance:")
print(f"   â€¢ Silhouette Score: {silhouette_score(X_scaled, df['cluster']):.3f}")
print(f"   â€¢ Davies-Bouldin Score: {davies_bouldin_score(X_scaled, df['cluster']):.3f}")

print(f"\nðŸ’¡ Key Findings:")
for cluster_id in range(optimal_k):
    cluster_size = len(df[df['cluster'] == cluster_id])
    cluster_grade = df[df['cluster'] == cluster_id]['mean_module_grade'].mean()
    print(f"\n   Cluster {cluster_id}:")
    print(f"      â€¢ Size: {cluster_size} students ({cluster_size/len(df)*100:.1f}%)")
    print(f"      â€¢ Avg Grade: {cluster_grade:.3f}")
    print(f"      â€¢ Performance: {'High' if cluster_grade >= 0.7 else 'Medium' if cluster_grade >= 0.4 else 'Low'}")

print(f"\nðŸ¤– Q-Learning Training:")
print(f"   â€¢ Episodes per Agent: {n_episodes}")
print(f"   â€¢ Learning Rate: {q_agents[0].lr}")
print(f"   â€¢ Discount Factor: {q_agents[0].gamma}")

print("\n" + "="*70)
print("âœ… ANALYSIS COMPLETE!")
print("="*70)
print("\nNext Steps:")
print("1. Review cluster characteristics and recommendations")
print("2. Implement optimal policies for each student group")
print("3. Monitor student progress and adjust strategies")
print("4. Collect more data for continuous improvement")

# %%
```
