### Tb theo cá»¥m mod Ä‘iá»ƒm


```python
import pandas as pd

# Äá»c file CSV
df = pd.read_csv('../data/udk_moodle_grades_course_670.csv')

# Lá»c ra cÃ¡c dÃ²ng cÃ³ itemtype = 'mod'
df_mod = df[df['itemtype'] == 'mod'][['userid', 'finalgrade', 'courseid', 'itemtype']]

# Lá»c ra cÃ¡c báº£n ghi há»£p lá»‡ (cÃ³ Ä‘iá»ƒm > -1)
df_valid_mod = df_mod[df_mod['finalgrade'] > -1]

# âœ… TÃ­nh Ä‘iá»ƒm trung bÃ¬nh tá»«ng há»c viÃªn (chá»‰ tÃ­nh mod)
avg_grade_per_user = (
    df_valid_mod
    .groupby('userid')['finalgrade']
    .mean()
    .reset_index()
    .rename(columns={'finalgrade': 'mean_mod_grade'})
)

avg_grade_per_user_count = (
    df_valid_mod
    .groupby('userid')['finalgrade']
    .count()
    .reset_index()
    .rename(columns={'finalgrade': 'mod_grade_count'})
)


# âœ… In káº¿t quáº£
print("ğŸ“Š Äiá»ƒm trung bÃ¬nh (chá»‰ cÃ¡c module) cá»§a tá»«ng há»c viÃªn:")
print(avg_grade_per_user)

# âœ… (Tuá»³ chá»n) Thá»‘ng kÃª thÃªm
print(f"\nSá»‘ lÆ°á»£ng há»c viÃªn cÃ³ Ä‘iá»ƒm mod: {len(avg_grade_per_user)}")
print(f"Äiá»ƒm trung bÃ¬nh toÃ n bá»™ há»c viÃªn: {avg_grade_per_user['mean_mod_grade'].mean():.2f}")
print(f"Äiá»ƒm trung vá»‹ toÃ n bá»™ há»c viÃªn: {avg_grade_per_user['mean_mod_grade'].median():.2f}")
print(f"Äiá»ƒm cao nháº¥t: {avg_grade_per_user['mean_mod_grade'].max()}")
print(f"Äiá»ƒm tháº¥p nháº¥t: {avg_grade_per_user['mean_mod_grade'].min()}")


print("\nğŸ“Š Sá»‘ lÆ°á»£ng module Ä‘Ã£ Ä‘Æ°á»£c cháº¥m Ä‘iá»ƒm cá»§a tá»«ng há»c viÃªn:")
print(avg_grade_per_user_count)
```

```python
import pandas as pd

# Äá»c file CSV Ä‘iá»ƒm
df = pd.read_csv('../data/udk_moodle_grades_course_670.csv')

# Lá»c itemtype = 'mod' vÃ  Ä‘iá»ƒm há»£p lá»‡
df_mod = df[df['itemtype'] == 'mod'][['userid', 'finalgrade']]
df_valid_mod = df_mod[df_mod['finalgrade'] > -1]

# TÃ­nh Ä‘iá»ƒm trung bÃ¬nh vÃ  sá»‘ module
avg_grade_per_user = (
    df_valid_mod
    .groupby('userid')['finalgrade']
    .mean()
    .reset_index()
    .rename(columns={'finalgrade': 'mean_mod_grade'})
)

avg_grade_per_user_count = (
    df_valid_mod
    .groupby('userid')['finalgrade']
    .count()
    .reset_index()
    .rename(columns={'finalgrade': 'mod_grade_count'})
)

# In káº¿t quáº£ thá»‘ng kÃª
print("ğŸ“Š Äiá»ƒm trung bÃ¬nh (chá»‰ cÃ¡c module) cá»§a tá»«ng há»c viÃªn:")
print(avg_grade_per_user)
print("\nğŸ“Š Sá»‘ lÆ°á»£ng module Ä‘Ã£ Ä‘Æ°á»£c cháº¥m Ä‘iá»ƒm cá»§a tá»«ng há»c viÃªn:")
print(avg_grade_per_user_count)

# XÃ¡c Ä‘á»‹nh user Ä‘Æ°á»£c giá»¯ láº¡i vÃ  bá»‹ lá»c
users_keep = avg_grade_per_user_count[avg_grade_per_user_count['mod_grade_count'] > 5]['userid'].tolist()
users_filtered = list(set(avg_grade_per_user_count['userid']) - set(users_keep))

print("\nâœ… User Ä‘Æ°á»£c giá»¯ láº¡i (Ä‘Ã£ lÃ m >5 module):")
print(users_keep)
print("\nğŸš« User bá»‹ lá»c (â‰¤5 module):")
print(users_filtered)

# Äá»c log vÃ  Ä‘áº¿m event cá»§a user bá»‹ lá»c
df_logs = pd.read_csv('../data/udk_moodle_log_course_670.csv')

df_logs_filtered = df_logs[df_logs['userid'].isin(users_filtered)]

event_count = (
    df_logs_filtered
    .groupby('userid')['id']
    .count()
    .reset_index()
    .rename(columns={'id': 'event_count'})
)

print("\nğŸ“Š Tá»•ng sá»‘ event cá»§a cÃ¡c user bá»‹ lá»c (â‰¤5 module):")
print(event_count)


# âœ… (Tuá»³ chá»n) Tá»•ng sá»‘ event chung cá»§a nhÃ³m bá»‹ lá»c
print(f"\nğŸ”¢ Tá»•ng event cá»§a toÃ n bá»™ nhÃ³m bá»‹ lá»c: {event_count['event_count'].sum()}")
```
