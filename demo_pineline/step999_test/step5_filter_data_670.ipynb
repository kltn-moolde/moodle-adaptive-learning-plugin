```python
array_filter = [9453, 2143, 7267, 8965, 9452, 9131, 9155, 3]

# ÄÆ°á»ng dáº«n file (tÆ°Æ¡ng Ä‘á»‘i Ä‘áº¿n thÆ° má»¥c demo_pineline/step999_test)
log_path = '../data/udk_moodle_log_course_670.csv'
grades_path = '../data/udk_moodle_grades_course_670.csv'

# ÄÆ°á»ng dáº«n Ä‘áº§u ra
out_log_path = '../data/udk_moodle_log_course_670.filtered.csv'
out_grades_path = '../data/udk_moodle_grades_course_670.filtered.csv'
report_path = '../data/filter_report_course_670.json'

import os
import json
import pandas as pd

# Kiá»ƒm tra file tá»“n táº¡i
missing_files = [p for p in (log_path, grades_path) if not os.path.exists(p)]
if missing_files:
    raise FileNotFoundError(f'Missing input files: {missing_files}')

# Äá»c dá»¯ liá»‡u
df_logs = pd.read_csv(log_path)
df_grades = pd.read_csv(grades_path)

# Kiá»ƒm tra cá»™t userid tá»“n táº¡i
for name, df in [('logs', df_logs), ('grades', df_grades)]:
    if 'userid' not in df.columns:
        raise ValueError(f'File {name} does not contain column "userid"')

print(f'Initial logs: {len(df_logs)}, grades: {len(df_grades)}')

df_logs_filtered = df_logs[~df_logs['userid'].isin(array_filter)]
df_grades_filtered = df_grades[~df_grades['userid'].isin(array_filter)]

# LÆ°u káº¿t quáº£
df_logs_filtered.to_csv(out_log_path, index=False)
df_grades_filtered.to_csv(out_grades_path, index=False)

print(f'Filtered logs: {len(df_logs_filtered)}, grades: {len(df_grades_filtered)}')
```
