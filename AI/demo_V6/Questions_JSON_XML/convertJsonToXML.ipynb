```python
import json
from lxml import etree as ET

def create_moodle_xml_lxml(json_data, filename="moodle_quiz.xml"):
    """
    Tạo file XML từ JSON theo chuẩn Moodle, dùng lxml và hỗ trợ CDATA.
    """
    quiz = ET.Element("quiz")

    for idx, q_data in enumerate(json_data["questions"], start=1):
        # Comment: Câu số mấy
        quiz.append(ET.Comment(f"Câu {idx}"))

        question = ET.SubElement(quiz, "question", type=q_data["question_type"])

        # <name>
        name = ET.SubElement(question, "name")
        ET.SubElement(name, "text").text = q_data["name"]

        # <questiontext>
        questiontext = ET.SubElement(question, "questiontext", format="html")
        qt_text = ET.SubElement(questiontext, "text")
        qt_text.text = ET.CDATA(q_data["question_text"])

        # Config
        ET.SubElement(question, "shuffleanswers").text = "0"
        ET.SubElement(question, "single").text = "true"
        ET.SubElement(question, "answernumbering").text = "abc"

        # Answers
        for a_data in q_data["answers"]:
            answer = ET.SubElement(question, "answer", fraction=str(a_data["fraction"]))
            ET.SubElement(answer, "text").text = a_data["text"]

            feedback = ET.SubElement(answer, "feedback")
            ET.SubElement(feedback, "text").text = a_data["feedback"]

        # Tags
        tags = ET.SubElement(question, "tags")
        tag = ET.SubElement(tags, "tag")
        ET.SubElement(tag, "text").text = q_data["difficulty"]

    # Ghi file XML
    tree = ET.ElementTree(quiz)
    tree.write(filename, pretty_print=True, xml_declaration=True, encoding="UTF-8")
    print(f"✅ File XML '{filename}' đã được tạo thành công.")
def create_moodle_xml_lxml1(json_file_path, filename="moodle_quiz.xml"):
    """
    Tạo file XML từ file JSON theo chuẩn Moodle, dùng lxml và hỗ trợ CDATA.
    """
    # Đọc dữ liệu JSON từ file
    with open(json_file_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)

    quiz = ET.Element("quiz")

    for idx, q_data in enumerate(json_data["questions"], start=1):
        quiz.append(ET.Comment(f"Câu {idx}"))

        question = ET.SubElement(quiz, "question", type=q_data["question_type"])

        # <name>
        name = ET.SubElement(question, "name")
        ET.SubElement(name, "text").text = q_data["name"]

        # <questiontext>
        questiontext = ET.SubElement(question, "questiontext", format="html")
        qt_text = ET.SubElement(questiontext, "text")
        qt_text.text = ET.CDATA(q_data["question_text"])

        # Config
        ET.SubElement(question, "shuffleanswers").text = "0"
        ET.SubElement(question, "single").text = "true"
        ET.SubElement(question, "answernumbering").text = "abc"

        # Answers
        for a_data in q_data["answers"]:
            answer = ET.SubElement(question, "answer", fraction=str(a_data["fraction"]))
            ET.SubElement(answer, "text").text = a_data["text"]

            feedback = ET.SubElement(answer, "feedback")
            ET.SubElement(feedback, "text").text = a_data["feedback"]

        # Tags
        tags = ET.SubElement(question, "tags")
        tag = ET.SubElement(tags, "tag")
        ET.SubElement(tag, "text").text = q_data["difficulty"]

    # Ghi file XML
    tree = ET.ElementTree(quiz)
    tree.write(filename, pretty_print=True, xml_declaration=True, encoding="UTF-8")
    print(f"✅ File XML '{filename}' đã được tạo thành công từ '{json_file_path}'.")

# Chạy hàm để tạo file XML
create_moodle_xml_lxml1(json_file_path="bai6.json", filename="moodle_quiz_bai6.xml")
```
