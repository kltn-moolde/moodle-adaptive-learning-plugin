```python
import requests
import json
import pandas as pd
from IPython.display import display, HTML

# Cấu hình Moodle API
moodle_url = "http://localhost:8100/webservice/rest/server.php"
api_token = "be7a58dda310c86830ec9a2031b40afb"

def call_moodle_api(wsfunction, params=None):
   
    default_params = {
        'wstoken': api_token,
        'wsfunction': wsfunction,
        'moodlewsrestformat': 'json'
    }
    
    if params:
        default_params.update(params)
    
    try:
        response = requests.get(moodle_url, params=default_params)
        if response.status_code != 200:
            print(f"❌ Lỗi HTTP: {response.status_code}")
            print("Nội dung phản hồi:", response.text)
            return None
        
        data = response.json()
        
        print(f"Phản hồi từ '{wsfunction}':")
        print(json.dumps(data, indent=2, ensure_ascii=False))
        
        return data
    
    except requests.exceptions.RequestException as e:
        print(f"❌ Lỗi khi gửi yêu cầu: {e}")
        return None
    except json.JSONDecodeError as e:
        print(f"❌ Lỗi phân tích JSON: {e}")
        print("Nội dung phản hồi:", response.text)
        return None

def mod_lesson_get_lesson(lessonid):
    return call_moodle_api('mod_lesson_get_lesson', {'lessonid': lessonid})


def mod_lesson_get_pages(lessonid):
    return call_moodle_api('mod_lesson_get_pages', {'lessonid': lessonid})

def mod_lesson_get_user_attempt(lessonid, userid, lessonattempt):
    return call_moodle_api('mod_lesson_get_user_attempt', {'lessonid': lessonid, 'userid': userid, 'lessonattempt': lessonattempt})

def mod_lesson_get_content_pages_viewed(lessonid, userid):
    return call_moodle_api('mod_lesson_get_content_pages_viewed', {'lessonid': lessonid, 'userid': userid})

def core_user_get_users_by_username(username_keyword=''):
    return call_moodle_api('core_user_get_users', {
        'criteria[0][key]': '',
        'criteria[0][value]': username_keyword
    })
    
def core_tag_get_tags(tagid):
    return call_moodle_api('core_tag_get_tags', {'tags': {'tagid': tagid}})


# Gọi các hàm để kiểm tra
if __name__ == "__main__":
    # mod_lesson_get_lesson(3)
    # core_tag_get_tags(6)
    mod_lesson_get_pages(5) 
    # mod_lesson_get_user_attempt(4, 4, 0)
    # mod_lesson_get_content_pages_viewed(4, 1)
    # core_user_get_users_by_username('u')
```
